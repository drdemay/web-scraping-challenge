## Collect NASA Mars News

```python
# Import dependencies
from splinter import Browser
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests
import time
```

```python
#create and update the executable path to the chromedrv
executable_path = {"executable_path": "C:\chromedrv\chromedriver.exe"}
browser = Browser("chrome", **executable_path, headless=False)
```

```python
# URL of page to be scraped
url = "https://mars.nasa.gov/news"
browser.visit(url)
```

```python
#Retrieve page with the requests module
response = requests.get(url)
```

```python
# Create BeautifulSoup object; parse with 'html.parser'
soup = bs(response.text, 'html.parser')
```

```python
# Examine the results, then determine element that contains sought info
print(soup.prettify())
```

```python
# results are returned as an iterable list
results = soup.find_all("div", class_="slide")
results
```

```python
# Create empty dictionary to save values from scraping
mars = {}
```

```python
# Loop through returned results
for result in results:
  
    # Retrieve the title for each news story and description paragraph
    title = result.find("div", class_='content_title')
    p = result.find("div", class_='rollover_description_inner')
    news_title = title.a.text
    news_text = p.text

    try:
        print("\n-----------------------------\n")
        print(news_title)
        print(news_text)
        
    except AttributeError as e:
        print(e)
```

```python
# add title and text to Mars dictionary
mars["News_Title"]=news_title
mars["News_Text"]=news_text
```

```python
# URL for image to collect
url = "https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars"
base_url = "https://www.jpl.nasa.gov"
browser.visit(url)
```

```python
#Retrieve page with the requests module
response = requests.get(url)
```

```python
soup = bs(response.text, 'html.parser')
```

```python
print(soup.prettify())
```

## Collect Mars Weather Information

```python
# URL of page to be scraped
url = "https://twitter.com/marswxreport?lang=en"
browser.visit(url)
```

## Gather Mars facts

```python
# We can use the read_html function in Pandas to automatically scrape any tabular data from a page.
# URL to scrape information from
url = "https://space-facts.com/mars/"
```

```python
# Read the tables on the page
mars_facts = pd.read_html(url)
mars_facts
```

```python
# Put the Mars facts table specified into a dataframe 
mars_facts_df = mars_facts[0]
mars_facts_df.columns = ["0", "1"]
mars_facts_df.head(10)
```

```python
# Generate HTML tables from the DataFrame.
html_table = mars_facts_df.to_html(index=False, header=False)
html_table
```

```python
# Strip unwanted newlines to clean up the table.
html_table.replace('\n', '')
```

```python
# add facts to the mars dictionary
mars["facts"] = html_table
```

```python
print(mars)
```

## Gather Mars Hemisphere Information

```python
# Visit the USGS Astrogeology site to obtain high resolution images for each of Mar's hemispheres.

# You will need to click each of the links to the hemispheres in order to find the image url to the full resolution image.

# Save both the image url string for the full resolution hemisphere image, and the Hemisphere title containing the 
# hemisphere name. Use a Python dictionary to store the data using the keys img_url and title.

# Append the dictionary with the image url string and the hemisphere title to a list. 
# This list will contain one dictionary for each hemisphere.
```

```python
# create dictionary for images
#hemisphere_images_urls = []
```

```python
# URL of page to be scraped
hemisphere_url = "https://web.archive.org/web/20181114171728/https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars"

# Retrieve page with the requests module
response = requests.get(url)
# Create BeautifulSoup object; parse with 'lxml'
soup = bs(response.text, 'lxml')
browser.visit(hemisphere_url)
time.sleep(2)
hemisphere_image_urls=[]
url_links = browser.find_by_css("a.product-item h3")
url_links
```

```python
print(soup.prettify())
```

```python
# Loop through to get hemisphere images and append to list

for i in range(len(url_links)):
        
    #create dictionary for each hemisphere
    hemisphere={}
    browser.find_by_css("a.product-item h3")[i].click()
    
    # get title
    hemisphere[title] = browser.find_by_css("h2.title").text

    
    # find the sample image scrape href
    sample_img = browser.find_link_by_text("Sample").first
    hemisphere["img_url "] = sample_img["href"]
    
    #append images 
    hemisphere_image_urls.append(hemisphere)
    
    browser.back()
```

```python
# add images to Mars dictionary 
mars["hemispheres"] = hemisphere_image_urls
```
